# spiral art using python

```python
 # Import the turtle module
import turtle
import random
#define colors to use
colors = ["red", "yellow", "green", "purple", "blue", "orange"]  
turtle.bgcolor("black") 

t = turtle.Turtle()     

#code for main spiral art
for x in range(200):
    # Set the pen color to the appropriate color from the list
    t.pencolor(colors[x % 6]) 
    
    t.width(x / 100 + 1)            
    t.forward(x)                   
    t.left(59) 
    
#code for the movements of the turltle
t.penup()
t.goto(300,0)
t.pendown()

#code for the small spiral arts
for x in range(50):
    # Set the pen color to the appropriate color from the list
    t.pencolor(colors[x % 6]) 
    
    t.width(x / 25 + 1)            
    t.forward(x)                   
    t.left(59) 
    
t.penup()
t.goto(-300,0)
t.pendown()

for x in range(50):
    # Set the pen color to the appropriate color from the list
    t.pencolor(colors[x % 6]) 
    
    t.width(x / 25 + 1)            
    t.forward(x)                   
    t.left(59)     
    
t.penup()
t.goto(0,300)
t.pendown()

for x in range(50):
    # Set the pen color to the appropriate color from the list
    t.pencolor(colors[x % 6]) 
    
    t.width(x / 25 + 1)            
    t.forward(x)                   
    t.left(59)     
    
t.penup()
t.goto(0,-300)
t.pendown()

for x in range(50):
    # Set the pen color to the appropriate color from the list
    t.pencolor(colors[x % 6]) 
    
    t.width(x / 25 + 1)            
    t.forward(x)                   
    t.left(59)    

t.penup()
t.goto(300,300)
t.pendown()

for x in range(50):
    # Set the pen color to the appropriate color from the list
    t.pencolor(colors[x % 6]) 
    
    t.width(x / 25 + 1)            
    t.forward(x)                   
    t.left(59)
    
t.penup()
t.goto(-300,300)
t.pendown()

for x in range(50):
    # Set the pen color to the appropriate color from the list
    t.pencolor(colors[x % 6]) 
    
    t.width(x / 25 + 1)            
    t.forward(x)                   
    t.left(59)
    
t.penup()
t.goto(-300,-300)
t.pendown()

for x in range(50):
    # Set the pen color to the appropriate color from the list
    t.pencolor(colors[x % 6]) 
    
    t.width(x / 25 + 1)            
    t.forward(x)                   
    t.left(59)
    
    
t.penup()
t.goto(300,-300)
t.pendown()

for x in range(50):
    # Set the pen color to the appropriate color from the list
    t.pencolor(colors[x % 6]) 
    
    t.width(x / 25 + 1)            
    t.forward(x)                   
    t.left(59)
    
    
# Hide the turtle cursor
t.hideturtle()          
turtle.done()         
```
